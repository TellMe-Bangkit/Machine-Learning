```python
import os
import shutil

# Direktori utama data
base_dir = './data/train'
sample_dir = './sampleData'

# Nama folder dan label
labels = list('ASHKUVTR') + ['del', 'space']

# Pastikan direktori sampleData ada
os.makedirs(sample_dir, exist_ok=True)

# Loop melalui setiap label
for label in labels:
    folder_path = os.path.join(base_dir, label)
    
    # Periksa apakah folder memiliki gambar
    if os.path.exists(folder_path) and os.listdir(folder_path):
        # Ambil satu file (gambar) pertama dari folder
        first_image = os.listdir(folder_path)[0]  # Ambil file pertama
        source_path = os.path.join(folder_path, first_image)
        
        # Tentukan nama file tujuan
        if label == 'del':
            dest_filename = 'del.jpg'
        elif label == 'space':
            dest_filename = 'space.jpg'
        else:
            dest_filename = f'{label}.jpg'

        dest_path = os.path.join(sample_dir, dest_filename)

        # Pindahkan file ke direktori sampleData
        shutil.move(source_path, dest_path)
        print(f"Pindahkan {source_path} ke {dest_path}")
    else:
        print(f"Tidak ada data di folder: {label}")
```

```python
import os
import cv2
import mediapipe as mp

# Inisialisasi MediaPipe Hands
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(
    static_image_mode=True,  # Karena gambar, gunakan mode statis
    max_num_hands=1,
    min_detection_confidence=0.3
)

# Direktori utama
base_dir = './data/train'

# Label yang diurutkan
labels = sorted(list('T'))

# Menyimpan semua path gambar
image_paths = []

# Mengumpulkan semua path gambar dari setiap folder label
for label in labels:
    folder_path = os.path.join(base_dir, label)
    if os.path.exists(folder_path):
        for filename in os.listdir(folder_path):
            if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_paths.append(os.path.join(folder_path, filename))

# Pastikan ada gambar untuk ditampilkan
if not image_paths:
    print("Tidak ada gambar yang ditemukan di folder.")
    exit()

# Inisialisasi indeks gambar
current_index = 0

# Menampilkan gambar satu per satu
while True:
    # Membaca gambar
    image_path = image_paths[current_index]
    image = cv2.imread(image_path)
    if image is None:
        print(f"Gagal membaca gambar: {image_path}")
        current_index = (current_index + 1) % len(image_paths)  # Lanjutkan ke gambar berikutnya
        continue

    # Convert gambar ke RGB untuk MediaPipe
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Proses gambar untuk mendeteksi tangan
    result = hands.process(image_rgb)

    # Jika ditemukan landmark, gambar pada gambar asli
    if result.multi_hand_landmarks:
        for hand_landmarks in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(
                image, 
                hand_landmarks, 
                mp_hands.HAND_CONNECTIONS
            )

    # Menampilkan path pada sudut kanan atas gambar
    text_size = cv2.getTextSize(image_path, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 2)[0]
    text_x = image.shape[1] - text_size[0] - 10  # Posisi teks pada sumbu x
    text_y = 30  # Posisi teks pada sumbu y
    cv2.putText(
        image,
        f"Path: {image_path}",
        (text_x, text_y),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.5,
        (0, 255, 0),
        2
    )

    # Tampilkan gambar
    cv2.imshow("Image Viewer with Landmarks", image)

    # Tunggu input pengguna
    key = cv2.waitKey(0) & 0xFF

    if key == ord('s'):  # Jika tombol 's' ditekan, lanjutkan ke gambar berikutnya
        current_index = (current_index + 1) % len(image_paths)
    elif key == ord('d'):  # Jika tombol 'd' ditekan, kembali ke gambar sebelumnya
        current_index = (current_index - 1 + len(image_paths)) % len(image_paths)
    elif key == ord('r'):  # Jika tombol 'r' ditekan, hapus gambar
        if os.path.exists(image_path):
            os.remove(image_path)  # Hapus file
            print(f"Gambar {image_path} telah dihapus.")
            image_paths.pop(current_index)  # Hapus path dari daftar
            # Periksa jika tidak ada lagi gambar
            if not image_paths:
                print("Tidak ada gambar yang tersisa.")
                break
            # Perbarui indeks untuk memastikan tidak keluar dari batas
            current_index = current_index % len(image_paths)
    elif key == ord('q'):  # Jika tombol 'q' ditekan, keluar dari loop
        break

# Tutup MediaPipe dan jendela OpenCV
hands.close()
cv2.destroyAllWindows()

```
